```python
unicorn_companies.City = unicorn_companies.City.fillna(method='ffill', axis=0)
```

```python
unicorn_companies['Select Investors']= unicorn_companies['Select Investors'].fillna(method='ffill', axis=0)
```

```python
unicorn_companies.loc[:,['Valuation', 'Funding']] = unicorn_companies.loc[:,['Valuation', 'Funding']].replace('[^0-9]','', regex=True).replace('',np.nan, regex=True).dropna(subset=['Valuation', 'Funding'])\
                  .astype(int)
```

```python
unicorn_companies.loc[:,['Valuation', 'Funding']] = unicorn_companies.loc[:,['Valuation', 'Funding']].fillna({
                 'Valuation':unicorn_companies.Valuation.mean(),
                 'Funding':unicorn_companies.Funding.mean()
})
```
